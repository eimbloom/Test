```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import openpyxl
import geoplot as gplt
import geoplot.crs as gcrs
```

```python
geojson = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
geojson
```

```python
df = pd.read_excel(r'C:\Users\ebloom\Desktop\Data_Extract_From_Adjusted_Net_Savings.xlsx',
                   usecols=["Country Name", "Country Code", "2017 [YR2017]", "2018 [YR2018]", "2019 [YR2019]"],
                  na_values=["0", ".."])
df.rename(columns = {"2017 [YR2017]" : "2017", "2018 [YR2018]" : "2018", "2019 [YR2019]" : "2019"}, inplace=True)
df = df.round({'2017' : 2, '2018' : 2, '2019' : 2})
df = df.drop(df.index[264:269])
df2 = df.drop(df.index[0:47])
df2
```

```python
merge = geojson.merge(df2, how='left', left_on='iso_a3', right_on='Country Code')
merge.dropna(subset=['2019'], inplace=True)
merge
```

```python
ax = gplt.webmap(merge, projection=gcrs.WebMercator(), zoom=3, figsize=(15,12))##provider=ctx.providers.CartoDB.Positron,)
gplt.choropleth(
    merge, hue='2019', projection=gcrs.WebMercator(),
    edgecolor='white', linewidth=1, alpha=0.7,
    cmap='viridis', legend=True, ax=ax)
plt.title("World natural resource rents 2019", fontsize=15)

##https://nbviewer.jupyter.org/github/allixender/30MapChallenge2020/blob/main/20/day-20.ipynb
```

```python
ax = gplt.cartogram(
    merge, scale='2019', hue='2019',
    legend=True, legend_var='hue',
    figsize=(15,12))
gplt.polyplot(merge, facecolor='lightgray', edgecolor='white', ax=ax)
```
